````python
import networkx as nx
G = nx.Graph()
G.add_nodes_from( ["Matt", "Leigh","Seth","Camille"] )
G.nodes()
````

````python
G.nodes()
````

````python
# Now add some weighted edges: correlation example, so 1.0 is maximum
G.add_edge("Matt","Leigh", weight=1.0)
G.add_edge("Matt","Seth", weight=0.888)
G.add_edge("Camille","Leigh", weight=0.567)
G.add_edge("Seth","Leigh", weight=0.667)
````

````python
# Used for layout later
pos=nx.spring_layout(G)
pos
````

Now it's time to make a pic of this graph. We could just include industry ID strings in the future, or whatever we wanted. I'm following the example ["weighted_graph" on the NetworkX docs](http://networkx.github.io/documentation/latest/examples/drawing/weighted_graph.html).

#### help(nx.draw_networkx_edges) (important points)
```
edge_color : color string, or array of floats
       Edge color. Can be a single color format string (default='r'),
       or a sequence of colors with the same length as edgelist.
       If numeric values are specified they will be mapped to
       colors using the edge_cmap and edge_vmin,edge_vmax parameters.

```

````python
# see help(nx.draw_networkx_edges) for more
nx.draw_networkx_nodes(G, pos, node_size=700)
nx.draw_networkx_edges(G, pos=pos, width=10, edge_cmap=get_cmap('Purples') )
#labels
nx.draw_networkx_labels(G, pos, font_size=20, font_family='serif')
#style points
plt.axis('off')
````

````python
colors = []
for el in G.edges():
    w = G.get_edge_data(el[0], el[1])['weight']
    colors.append( w )
````
